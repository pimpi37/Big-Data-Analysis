## 데이터 준비

```python
# 결측치 확인
titanic.isnull()
```

```python
# 결측치 확인
titanic.isnull().sum()
```

```python
# age 열의 결측치(값)을 중앙값으로 치환
# fillna(): 빈칸 채우기, median(): 중앙값 
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
```

```python
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
```

```python
# titanic dataframe에 새로운 열 'age2'추가 
titanic['age2'] = titanic['age'].apply(category_age)
```

```python
titanic.head(2)
```

```python
# 'sex' 열의 male을 1로, female을 0으로 치환
titanic['sex'] = titanic['sex'].map({'male':1, 'female': 0})
```

```python
titanic.head(2)
```

```python
# 가족의 수를 구하여 family열을 추가(sibsp : 함께 탑승한 형제자매와 배우자 수, parch: 함께 탑승한 부모/자식 수)
# 끝에 +1은 본인을 합한 것.
titanic['family'] = titanic['sibsp'] + titanic['parch'] + 1
```
