```python
# 단순회귀분석: 종속 변수(ex:아이스크림 판매량)에 영향을 미치는 독립변수(ex:날씨,시간,지역)가 1개
# 다중회귀분석: 종속 변수에 영향을 미치는 독립변수가 여러 개
```

```python
wine.head()
```

```python
# '종속변수 ~ 독립변수 + 독립변수+....'
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
# 선형회귀분석
regression_result = ols(Rformula, data = wine).fit()
```

```python
regression_result.summary()
# R-squared : 0.3 - 독립변수가 종속변수의 30% 정도를 설명한다.
# R-squared는 독립변수가 종속변수를 얼마나 잘 설명하는 지를 나타냅니다. R-squared는 0과 1 사이 값을 가집니다
# 일반적으로는 20%가 넘어야 한다고 봄.
```
